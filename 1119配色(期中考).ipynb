```
ggplot(NBA1011,aes(x=FieldGoalsAttempted,y=TotalPoints))+
geom_col()+geom_point()+geom_smooth()
```

```
qplot(x=FieldGoalsAttempted,y=TotalPoints,data=NBA1011,
geom=c("point","smooth"),color=ThreesMade)
```

取NBA1011的資料，以FieldGoalsAttempted為x軸、TotalPoints為y軸，作散佈圖，將ThreesMade著色，最後畫出得分的漸進線

```
NBA1011%>%count(GamesPlayed)
```

```
NBA1011%>%
count(GamesPlayed)%>%
ggplot(aes(x=GamesPlayed,y=n,fill=GamesPlayed))+geom_col()+
geom_text(aes(label=n),vjust=-0.25)+
labs(x="GamesPlayed",y="times")
```

```
qplot(GamesPlayed,data=NBA1011,fill=Team,binwidth=1)
```

取NBA1011的資料，以GamesPlayed為x軸作長條圖，每個直條寬度為5，並將各Team標以不同顏色

```
ggplot(NBA1011,aes(x=TotalMinutesPlayed,y=FieldGoalsMade))+geom_col()+geom_smooth()
```

```
qplot(TotalMinutesPlayed,FieldGoalsMade,data=NBA1011,
color=ThreesMade,geom=c("point","smooth"),facets=.~Position,)
```

取NBA1011的資料，以TotalMinutesPlayed為x軸，以FieldGoalsMade為y軸，作各Position的散佈圖，且以直向呈現，並將ThreesMade著色，最後畫出漸進線

```
dataframe<-data.frame(x=factor(rep(letters[1:10],each=2)),y=rnorm(100))
```

取NBA1011的資料，將data.frame(x_axis=factor(rep(letters[1:10],each=2)),y_axis=rnorm(100))取名為dataframe，設x的factor為rep(letters[1:10],each=2)，設y的factor為rnorm(100)

```
ggplot(dataframe,aes(x=x,y=y))+geom_col()+geom_boxplot()
```
