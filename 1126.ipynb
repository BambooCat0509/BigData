```
data<-read.csv("Employee_Salary.csv")
```

```
str(data)
```

```
data
```

```
data[order(data$受僱員工人數,decreasing=T),][1:5,]
data[order(data$總薪資,decreasing=T),][1:5,]
data[order(data$經常性薪資,decreasing=T),][1:5,]
data[order(data$非經常性薪資,decreasing=T),][1:5,]
```

```
ggplot(data,aes(x=受僱員工人數,y=職類別,fill=行業別))+geom_col()+labs(title="不同行業的各職類別僱用員工人數比較")
ggplot(data,aes(x=總薪資,y=職類別,color=行業別))+geom_point()+labs(title="不同行業的各職類別總薪資比較")
ggplot(data,aes(x=總薪資,y=受僱員工人數,color=行業別))+geom_line()+labs(title="不同行業的受僱員工人數和總薪資比較")
```

```
ggplot(data,aes(x=總薪資,y=經常性薪資))+geom_point()+facet_grid(.~行業別)+geom_smooth()
ggplot(data,aes(x=總薪資,y=非經常性薪資))+geom_point()+facet_grid(.~行業別)+geom_smooth()
ggplot(data,aes(x=經常性薪資,y=非經常性薪資))+geom_point()+facet_grid(.~行業別)+geom_smooth()
```

```
a<-subset(data,行業別=="礦業及土石採取業各業")
```

```
a[order(a$受僱員工人數,decreasing=T),][1:10,]
a[order(a$總薪資,decreasing=T),][1:10,]
a[order(a$經常性薪資,decreasing=T),][1:10,]
a[order(a$非經常性薪資,decreasing=T),][1:10,]
```

```
an<-fct_reorder(factor(a$職類別),a$受僱員工人數)
a[order(a$受僱員工人數,decreasing=T),]
ats<-fct_reorder(factor(a$職類別),a$總薪資)
a[order(a$總薪資,decreasing=T),]
acs<-fct_reorder(factor(a$職類別),a$經常性薪資)
a[order(a$經常性薪資,decreasing=T),]
ans<-fct_reorder(factor(a$職類別),a$非經常性薪資)
a[order(a$非經常性薪資,decreasing=T),]
```

```
ggplot(a,aes(x=總薪資,y=ats))+geom_col()+geom_text(aes(label=總薪資),vjust=0.5)+labs(title="不同職類別的總薪資比較")
ggplot(a,aes(x=受僱員工人數,y=an))+geom_col()+geom_text(aes(label=受僱員工人數),vjust=0.5)+labs(title="不同職類別的受僱員工人數比較")
ggplot(a,aes(x=經常性薪資,y=acs))+geom_col()+geom_text(aes(label=經常性薪資),vjust=0.5)+labs(title="不同職類別的經常性薪資比較")
ggplot(a,aes(x=非經常性薪資,y=ans))+geom_col()+geom_text(aes(label=非經常性薪資),vjust=0.5)+labs(title="不同職類別的非經常性薪資比較")
```

```
ggplot(a,aes(x=ats,y=非經常性薪資,))+geom_point()+labs(title="不同總薪資的非經常性薪資比較")
ggplot(a,aes(x=ats,y=經常性薪資))+geom_point()+labs(title="不同總薪資的經常性薪資比較")
ggplot(a,aes(x=acs,y=非經常性薪資))+geom_point()+labs(title="不同經常性薪資的非經常性薪資比較")
```
