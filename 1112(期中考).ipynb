```
NBA1011<-fetch_NBAPlayerStatistics("10-11")
```

將NBAPlayerStatistics("10-11")的資料名稱設為NBA1011

```
str(NBA1011)
```

讀取並顯示NBA1011的資料

```
TotalPoints<-NBA1011[order(NBA1011$TotalPoints,decreasing=T),]
```

設NBA1011[order(NBA1011$TotalPoints,decreasing=T),]名稱為TotalPoints

```
TotalPoints[1:20,"Name"]
```

顯示TotalPoints中，得分前20高的人Name

```
TotalPoints[1:20,"TotalPoints"]
```

顯示TotalPoints中，得分前20高的TotalPoints

```
subset(NBA1011,Team=="GSW")
```

搜索NBA1011中，Team為GSW的所有資料並顯示

```
NBA1011[grepl("Curry",NBA1011$Name,),]
```

找出NBA1011中，所有Name有Curry的資料，並顯示出來

```
qplot(FieldGoalsAttempted,TotalPoints,data=NBA1011,
color=ThreesMade,geom=c("point","smooth"))
```

取NBA1011的資料，以FieldGoalsAttempted為x軸、TotalPoints為y軸，作散佈圖，將ThreesMade著色，最後畫出得分的漸進線

```
qplot(GamesPlayed,data=NBA1011,fill=Team,binwidth=5)
```

取NBA1011的資料，以GamesPlayed為x軸作長條圖，每個直條寬度為5，並將各Team標以不同顏色

```
qplot(TotalMinutesPlayed,FieldGoalsMade,data=NBA1011,
color=ThreesMade,geom=c("point","smooth"),facets=.~Position,)
```

取NBA1011的資料，以TotalMinutesPlayed為x軸，以FieldGoalsMade為y軸，作各Position的散佈圖，且以直向呈現，並將ThreesMade著色，最後畫出漸進線

```
dataframe<-data.frame(x_axis=factor(rep(letters[1:10],each=2)),y_axis=rnorm(100))
```

取NBA1011的資料，將data.frame(x_axis=factor(rep(letters[1:10],each=2)),y_axis=rnorm(100))取名為dataframe，設x的factor為rep(letters[1:10],each=2)，設y的factor為rnorm(100)

```
ggplot(dataframe,aes(x=x_axis,y=y_axis))+geom_boxplot()
```

取dataframe的資料，設x_axis為x軸，y_axis為y軸，作盒狀圖
