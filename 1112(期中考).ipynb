```
NBA1011<-fetch_NBAPlayerStatistics("10-11")
```

將NBAPlayerStatistics("10-11")的資料名稱設為NBA1011

```
str(NBA1011)
```

讀取NBA1011的資料

```
TotalPoints<-NBA1011[order(NBA1011$TotalPoints,decreasing=T),]
```

設NBA1011[order(NBA1011$TotalPoints,decreasing=T),]名稱為TotalPoints

```
TotalPoints[1:20,"Name"]
```

顯示TotalPoints中，得分前20高的人名

```
TotalPoints[1:20,"TotalPoints"]
```

顯示TotalPoints中，得分前20高的分數

```
subset(NBA1011,Team=="GSW")
```

搜索NBA1011中，Team為GSW的所有資料並顯示

```
NBA1011[grepl("Curry",NBA1011$Name,),]
```

找出所有Name中有Curry的資料

```
qplot(FieldGoalsAttempted,TotalPoints,data=NBA1011,
color=ThreesMade,geom=c("point","smooth"))
```

以FieldGoalsAttempted為x軸，以TotalPoints為y軸做散佈圖，並且將ThreesMade著色，最後畫出得分的漸進線

```
qplot(TotalMinutesPlayed,data=NBA1011,fill=Team,binwidth=100)
```

以TotalMinutesPlayed為x軸做長條圖，每個直條寬度為10，各隊標以不同顏色，

```
qplot(FieldGoalsAttempted,TotalPoints,data=NBA1011,
color=ThreesMade,geom=c("point","smooth"),facets=.~Position,)
```

以FieldGoalsAttempted為x軸，以TotalMinutesPlayed為y軸，做各個球場的散佈圖，並將ThreesMade著色，最後畫出漸進線

```
df<-data.frame(gp=factor(rep(letters[1:4],each=10)),y=rnorm(20))
```

將data.frame(gp=factor(rep(letters[1:4],each=10)),y=rnorm(20))取名為df，設gp的參數為rep(letters[1:4],each=10)，設y的參數為rnorm(20)

```
ggplot(df,aes(x=gp,y=y))+geom_boxplot()
```

取df的資料，取gp為x軸，y為y軸，做盒狀圖
